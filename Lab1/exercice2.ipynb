```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
image = cv2.imread('fondvert.png')  
```

```python
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)# Convert the image from BGR to RGB format (since OpenCV loads images in BGR format)
plt.imshow(image_copy)
```

```python
# Define the range for the color green (used in the background)
lower_green = np.array([0, 150, 0], dtype=np.uint8)
upper_green = np.array([125, 255, 120], dtype=np.uint8) 
```

```python
#Create a mask to identify the green regions in the image
mask = cv2.inRange(image_copy, lower_green, upper_green)
plt.imshow(mask, cmap='gray')
```

```python
# Change the green background to blue
background_change_blue = np.copy(image_copy)  # Create another copy of the original image
background_change_blue[mask != 0] = [0, 0, 255]  # Change all the green background pixels to blue (BGR value for blue)
plt.imshow(background_change_blue)
```
